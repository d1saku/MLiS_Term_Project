```python
import numpy as np

x = df.drop(["Sample_code_number"], axis=1).dropna()
classes = x["Class"]
X = x.drop(["Class"], axis=1)

X = X.values
X = X.astype(np.float64)

cov_matrix = np.cov(X.T)  # X is your data matrix, shape (n_samples, n_features)

# Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Feature loadings
loadings = eigenvectors * np.sqrt(eigenvalues)

# Optionally, organize into a readable format
feature_names = [f"Feature_{i+1}" for i in range(X.shape[1])]
loadings_df = pd.DataFrame(loadings, index=feature_names, columns=[f"PC{i+1}" for i in range(len(eigenvalues))])

print("Feature Loadings:")
print(loadings_df)
```

```python
from sklearn.decomposition import PCA

x = df.drop(["Sample_code_number"], axis=1).dropna()
classes = x["Class"]
x = x.drop(["Class"], axis=1)

pca = PCA(n_components=2)
pca.fit(x)

x_pca = pca.transform(x)

plt.figure(figsize=(8, 6))
plt.scatter(x_pca[:, 0]*-1, x_pca[:, 1], c=classes)
```

```python
from pca import PCA

x = df.drop(["Sample_code_number"], axis=1).dropna()
classes = x["Class"]
x = x.drop(["Class"], axis=1)

pca_custom = PCA(n_components=2)
pca_custom.fit(x)

x_pca = pca_custom.transform(x)

classes = classes.to_numpy()

class_2 = x_pca[classes == 2]
class_4 = x_pca[classes == 4]

plt.figure(figsize=(8, 6))
plt.scatter(class_2[:, 0], class_2[:, 1], color='blue', label='Class 2', edgecolor='k')
plt.scatter(class_4[:, 0], class_4[:, 1], color='red', label='Class 4', edgecolor='k')

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Results with Class Labels")
plt.legend(loc="best")


```
